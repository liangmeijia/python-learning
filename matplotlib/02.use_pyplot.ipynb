```python
x = np.arange(-2.0,2.0,0.1)
plt.plot(x,x,'r--',x,x**2,'bs',x,x**3,'g^')
```

使用关键字字符画图

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0,50,50),
        'd': np.random.randn(50)}
data['b'] = data['a']+ 10 * np.random.randn(50)
data['d']= np.abs(data['d'])*100
#print(data)
plt.scatter('a','b',s='c',c='d',data=data) #当data是结构化数据（列表、dataframe）时，可以使用关键字符取数据
```

使用分类变量画图

```python
data = ['group_a','group_b','group_c','group_d']
values =[20,40,30,80]

plt.figure(figsize=(9,3))
plt.subplot(131) #有1行3列的画布，此时位于第一个画布上
plt.bar(data,values)
plt.subplot(132)
plt.scatter(data,values)
plt.subplot(133)
plt.plot(data,values)
plt.suptitle('test plotting')
```

多画板

```python
def fun(x):
    return x**2

x = np.arange(0,5,0.1)
y = np.arange(0,5,0.02)

plt.figure()
plt.subplot(321)
plt.plot(x,fun(x),'ro',x,fun(x),'-')
plt.subplot(322)
plt.plot(x,np.cos(2*np.pi*x),aa=False,c = 'yellow',alpha=0.5)
plt.subplot(323)
plt.plot(x,np.sin(2*np.pi*x))
plt.subplot(324)
plt.plot(x,np.cos(2*np.pi*x),aa=False,c = 'b',alpha=0.5)
```

多画板figure之间的次序

```python
x = np.linspace(0,5,10)

plt.figure(1)
plt.suptitle('figure 1')
plt.subplot(121)
plt.plot(x,x**2,'ro')
plt.subplot(122)
plt.plot(x,x*2,'ko')

plt.figure(2)
plt.suptitle('figure 2')
plt.subplot(121)
plt.plot(x,x**3,'bd')
plt.subplot(122)
plt.plot(x,x**4,'rs')
```
