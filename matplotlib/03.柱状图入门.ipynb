```python
import matplotlib.pyplot as plt
import numpy as np
```

常规柱状图显示数据

```python
fruits = ['apple','banana','cherry'] #x轴是列表或元组形式都行
price = [4,5,15]
colors = ['b','g','r']

fig,ax = plt.subplots()
b = ax.bar(fruits,price,width=0.4,color=colors,label = fruits)
ax.bar_label(b,label_type='center')
ax.legend()
```

多组数据同时展示

```python
species =['Adelie','chinstrap','Gentoo']
sex_counts={
    'male': np.array([17,22,33]),
    'female': np.array([19,12,43])
}
bottom = np.zeros(3)

fig,ax = plt.subplots()
for sex,sex_count in sex_counts.items():
    b = ax.bar(species,sex_count,label= sex,width=0.4,bottom=bottom)
    bottom += sex_count 
    plt.bar_label(b,label_type='center')
ax.legend()
```

```python
species =['Adelie','Gentoo','Chinstrap'] 
means= {'Bill Depth':(18.35,18.43, 14.98),
        'Bill Length':(38.79, 48.83, 47.50),
        'Flipper Length':(189.95,195.82,217.19)}
x = np.arange(len(species))
offset = 0
wid = 0.25

fig,ax = plt.subplots()
for key,values in means.items():
    bars = ax.bar(x+offset,values,label=key,width=wid)
    ax.bar_label(bars,label_type='edge')
    ax.set_xticks(x+wid,species)
    offset+=wid
ax.legend(loc='upper left',ncol=3 )
ax.set_ylim([0,300])
```

横向带误差的柱状图

```python
people =('Tom','Dick','Slim','Jim','Harry')
y_pos = np.arange(len(people))
x = 2+10*np.random.rand(len(people))
print(x)
error = np.random.rand(len(people))
print(error)
fig,ax = plt.subplots()
#画横向柱状图
bars = ax.barh(y_pos,x,height=0.4,color='b',xerr=error,align='center')
ax.bar_label(bars,fmt='%.2f')
ax.set_yticks(y_pos,labels=people)
ax.set_xlim(0,14)
```

标注误差值

```python
people =('Tom','Dick','Slim','Jim','Harry')
y_pos = np.arange(len(people))
x = 2+10*np.random.rand(len(people))
print(x)
error = np.random.rand(len(people))
print(error)
fig,ax = plt.subplots()
#画横向柱状图 
bars = ax.barh(y_pos,x,height=0.4,color='b',xerr=error,align='center')
#自定义bar labels的值
ax.bar_label(bars,labels=[f'+{e:.2f}' for e in error] ,fmt='%.2f',
             fontsize='xx-small',color='r',padding=6)
ax.set_yticks(y_pos,labels=people)
ax.set_xlim(right=14)
```

matplotlib的内置画图风格
