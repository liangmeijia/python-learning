```python
import matplotlib.pyplot as plt
import numpy as np
from PIL.ImageOps import scale
```

茎图stem

堆叠图stackplot

箱型图boxplot

```python
x1 = [-2,-1,0,0,1,1,2,4,4,6,7,8,9,9,13,66,11,12,12,13,13,13,15,16,16,18,22,22,24,23,55]
x2 = [2,2,3,5,5,6,1,1,4,6,7,1,1,2,3,4,5,6,78]
x3 = [44,44,22,2,21,44,67,88,76,65,78,89,99,76,55,44]

fig,ax = plt.subplots()
ax.boxplot([x1,x2,x3],widths=0.2,positions=[2,4,6],showfliers=True,showmeans=True,
           medianprops=dict(linewidth=1,color='r',alpha=0.5),
            boxprops=dict(linewidth=2,color='y',alpha=0.5),
           capprops=dict(linewidth=1,color='b',alpha=0.5),
           whiskerprops=dict(linewidth=1,color='g',alpha=0.5),)
```

小提琴图

```python
x = np.random.normal([1,0,1],[0.75,0.25,1],[200,3])

fig,ax = plt.subplots()
v = ax.violinplot(x,positions=[2,4,6],widths=2,
              showmeans=True,showextrema=True,showmedians=True)

for i in v['bodies']:
    i.set_facecolor('green')
    i.set_edgecolor('green')
    i.set_alpha(0.5)
```

饼图pie

```python
scores = [66,59,90,88,86,100]
colors = plt.get_cmap('Reds')(np.linspace(0,1,len(scores)))

fig,ax = plt.subplots()
ax.pie(scores,radius=1,center=(4,4), #自动计算scores每一个值所占的比例
       wedgeprops=dict(edgecolor='white',linewidth=2),
       colors=colors
       )
```

等值线tricontour

```python
np.random.seed(1)
x = np.random.uniform(-3,3,1000)
y = np.random.uniform(-3,3,1000)
z = (1-x/2+x**5+y**2)*np.exp(-x**2-y**2)
levels = np.linspace(z.min(),z.max(),7)

plt.style.use('default')
fig,ax = plt.subplots()
#t = ax.tricontour(x,y,z,levels=levels) #在坐标(x,y)上的z相等，就画在一条线上；levels可以直接指定整数，代表画几条等值线
t = ax.tricontourf(x,y,z,levels=levels,cmap='rainbow') #带填充的等值线
plt.colorbar(t)
ax.scatter(x,y,s=10*z,c='gray')
```

三角网格

```python
np.random.seed(1)
x = np.random.uniform(-3,3,1000)
y = np.random.uniform(-3,3,1000)
z = (1-x/2+x**5+y**2)*np.exp(-x**2-y**2)

plt.style.use('default')
fig,ax = plt.subplots()

ax.plot(x,y,'o',markersize=1,color='gray')
t = ax.tripcolor(x,y,z,shading='flat') # 根据三角形的三个顶点的颜色值进行线性插值，以确定三角形内部每一点的颜色
plt.colorbar(t)
```

矩形网格pcolormesh

```python
x,y = np.meshgrid(np.linspace(-3,3,256),np.linspace(-3,3,256))
z = (1-x/2+x**5+y**2)*np.exp(-x**2-y**2)

fig,ax = plt.subplots()
p = ax.pcolormesh(x,y,z,cmap='rainbow')
plt.colorbar(p)
```

imshow

```python
x,y = np.meshgrid(np.linspace(-3,3,256),np.linspace(-3,3,256))
z = (1-x/2+x**5+y**2)*np.exp(-x**2-y**2)

fig,ax = plt.subplots()
ax.imshow(z,cmap='rainbow') #imshow直接将二维数据展示为矩形网格形式，注意此方式的坐标轴取值和pcolormesh方式的不同
```

矢量图quiver

```python
x,y = np.meshgrid(np.linspace(-4,4,6),np.linspace(-4,4,6))
p = x+y
q = x-y

fig,ax = plt.subplots()
ax.quiver(x,y,p,q,width=0.0045,scale=95,color='r') #p是x轴的偏移量, q是y轴的偏移量,scale是缩小的倍数
ax.set(xlim=(-5,5),ylim=(-5,5))
```

三维图形

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
z = x*y

fig,ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.stem(x,y,z)
```

```python
x,y,z = np.meshgrid([-1,0,1],[-1,0,1],[-1,0,1])
p = (x+y)/3
q = (x-y)/3
w = z*0
fig,ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.quiver(x,y,z,p,q,w)
```

```python
n=100
xs = np.linspace(0,1,n)
ys = np.sin(xs*np.pi*6)
zs = np.cos(xs*np.pi*6)

fig,ax = plt.subplots(subplot_kw=dict(projection='3d'))
ax.plot(xs,ys,zs,marker='o',color='r',markersize=2)
```
